### Поиск аномалий 

Проанализируем данные по объемам продаж в разных регионах с целью поиска выбросов и аномалий.

Для визуализации распределения и поиска аномалий подготовим функцию `draw_boxplot_hist`, которая выведет показатели описательной статистики, диаграмму размаха и гистограмму для столбца.

Для более детального рассмотрения распределения функция также выведет боксплот для определенного диапазона значений.

```python
def draw_boxplot_hist(
    df, 
    input_column,
    limit
):
    
# Показатели описательной статистики
    if limit==0:
        display(
            df[input_column].describe(
                include='all'
            ))

# Диаграмма размаха 
    flierprops = dict(
        marker='o', 
        markerfacecolor='None', 
        markersize=3,  
        markeredgecolor='black'
    )
    sns.set(
        rc={
            'figure.figsize':(16,1.5)
        })
    ax = sns.boxplot(
        x=df[input_column],
        showmeans=True,
        color=COLOR,
        width=.4,
        meanprops={
            "marker":"o",
            "markerfacecolor":"white", 
            "markeredgecolor":"black",
            "markersize":"8"},
        flierprops = flierprops
    )
    plt.title(
        'Диаграмма размаха для распределения значений {0}'.format(
            input_column.upper()), 
        fontsize=15
    )
    if limit!=0:
        if df[input_column].min()>0:
            plt.xlim(
                df[input_column].min()-.1*df[input_column].min(), 
                limit
            );
        else:
            plt.xlim(
                df[input_column].min()-.05*(
                    limit-df[input_column].min()
                ),
                limit
            );
    
# Кривая плотности с гисmoграммой  
    if limit==0:
        fig= plt.subplots(
            figsize=(16,3)
        )
        sns.distplot(
            df[input_column], 
            color=COLOR
        ).set_title(
            'Гистограмма размаха для распределения значений {0}'.format(
                input_column.upper()), 
            fontsize=15
        );
```

#### Общий объем продаж

Выведем визуализацию для столбца `total_sales` с помощью функции `draw_boxplot_hist`.

На графиках представляется, что данные сильно сконцентрированы около нуля, однако такая картина, вероятнее всего, возникает из-за выбросов, расположенных далеко за границами "усов" боксполота.

Рассчитаем, каков процент данных, попадающих в эту категорию. Для этого подготовим функцию `calculate_outliers`.

```python
def calculate_outliers(column, limit):
    print(
        'Процент аномальных данных в столбце:',
        ((
            data[
                data[column] > limit
            ][column].count() * 100
        )/ data[column].count()).round(4),
        '%'
    )
    display(data[data[column] > limit].head())
```

#### Объемы продаж по топ-5 платформам

Найдем платформы с самым большим суммарным объемом продаж, сгруппировав данные по платформам и просуммировав значения `total_sales` за все годы.

```python
def group_top(df):
    result=df.groupby([
        'platform'
    ])[['total_sales']].agg(
        'sum'
    ).sort_values(
        by='total_sales', 
        ascending = False
    ).reset_index().head(10)
    return result
```

```python
group_top(data)
```

```python
def top_platforms_to_top_games(df,top):
    list = group_top(df)[
        'platform'
    ].head(top).to_list()
    
    result = df.query('platform in @list')
    return result
```

#### Подготовка функций для анализа продаж

Для анализа продаж по каждой платформе в разных регионах подготовим несколько функций:

* `draw_platform_per_country` - визуализация распределения продаж по каждому из регионов
* `sum_total_sales` - сумма общих продаж в мире по годам 
* `draw_graph_per_platform` - визуализация распределения продаж по определенному региону
* `sum_sales_per_region` - суммарные продажи по каждому из регионов за весь рассматриваемый период

```python
def draw_platform_per_country(platform_name):
    columns=[[
        'na_sales', 
        'eu_sales', 
        'jp_sales', 
        'other_sales', 
        'total_sales'
    ]]
    
    for column in columns:
        fig, ax = plt.subplots(figsize=(15, 5))
        data[
            data['platform']==platform_name
        ].groupby('year_of_release')[column].sum().plot(
            alpha=0.75, 
            style=STYLE,
            ax=ax
        )
        
    plt.title(
        'Распределение общего объема продаж по платформе {0} по регионам'.format(
            platform_name), 
        fontsize=15
    )
    plt.xlabel(
        'Год выпуска', 
        fontsize=12
    )
    plt.ylabel(
        'Общий объем продаж (млн копий)', 
        fontsize=12
    )

    plt.xticks(np.arange(
        data[
            data['platform']==platform_name
        ]['year_of_release'].min(), 
        data[
            data['platform']==platform_name
        ]['year_of_release'].max()+1, 
        1
    ))
    
    plt.grid(axis='y', alpha=0.75)
    plt.show();
```

```python
def sum_total_sales(platform):
    display(
        data[
            data['platform']==platform
        ].groupby(
            'year_of_release'
        )['total_sales'].sum())
```

```python
def draw_graph_per_platform(platform_name,region):

    plt.figure(figsize=[15,2.5])
    data[
        data['platform']==platform_name
    ].groupby(
        'year_of_release'
    )[region].sum().plot(
        alpha=0.75, 
        style=STYLE,
        color=COLOR
    )
    plt.title(
        'Распределение общего объема продаж по платформе {0} в {1}'.format(
            platform_name, 
            region
        ),
        fontsize=15
    )
    plt.xlabel(
        'Год выпуска', 
        fontsize=12
    )
    plt.ylabel(
        'Общий объем продаж (млн копий)', 
        fontsize=12
    )
    plt.xticks(np.arange(
        data[
            data['platform']==platform_name
        ]['year_of_release'].min(), 
        data[
            data['platform']==platform_name
        ]['year_of_release'].max()+1, 
        1
    ))
    plt.grid(axis='y', alpha=0.75)
    plt.show()
```

```python
def sum_sales_per_region(platform):
    display(
        data[
            data['platform']==platform
        ][[
            'na_sales', 
            'eu_sales', 
            'jp_sales', 
            'other_sales', 
            'total_sales'
        ]].sum())
```

###  Общее распределение игр по жанрам

Сгруппируем данные по жанру чтобы оценить распределение суммарных объемов продаж по всем регионам. Для этого подготовим функцию `group_agg_func`.

```python
def group_agg_func(agg_func):
    result = pd.DataFrame({
        agg_func: data_actual.groupby([
            'genre'
        ])['total_sales'].agg(agg_func)
        }).reset_index()
    
    result=result.sort_values(
        by=agg_func,
        ascending = False
    ).reset_index()
    
    del(result['index'])
    result_2 = result[:7].copy()   
    new_row = pd.DataFrame(
        data = {
            'genre' : ['Other'],
            agg_func : [
                result[agg_func][7:].sum()
                ]})
    result_2 = pd.concat([
        result_2, 
        new_row
    ])
 
    return result_2
```

Также подготовим функцию `draw_piechart_solid` для построения круговой диаграммы.

```python
def draw_piechart_solid(input_data,agg_func):
    labels = input_data['genre']
    sizes = input_data[agg_func]
    fig1, ax1 = plt.subplots(figsize=(7,7))
    cmap=plt.get_cmap('PRGn_r')
    colors = cmap(
        np.linspace(
            0.05,
            0.8, 
            len(
                input_data['genre']
            )))
    textprops = {
        "fontsize":13
    }
    ax1.pie(
        sizes, 
        labels=labels, 
        autopct='%1.1f%%',
        colors=colors,
        startangle=90,
        textprops = textprops
    )
    ax1.axis('equal') 
    plt.title(
        'Доля продаж по жанрам по агрегирующей функции "{0}"'.format(agg_func), 
        fontsize = 15
    )
    plt.show()
```

### Подготовка функций для анализа портрета пользователя

Для анализа портрета пользователя для каждого из регионов подготовим функции `group_country`, которая сгруппирует, просуммирует данные и `draw_piechart`, которая построит соответствующую круговую диаграмму.

```python
def group_country(df,group,region,top):
    result = pd.DataFrame({
        'sum': df.groupby([
            group
        ])[region].agg('sum').round(1)
        }).reset_index()
    
    result=result.sort_values(
        by='sum',
        ascending = False
    ).reset_index()
    
    del(result['index'])
    result_2 = result[:top].copy()
    
    new_row = pd.DataFrame(
        data = {
            group : ['Other'],
        'sum' : [result['sum'][top:].sum()
                ]})

    result_2 = pd.concat([
        result_2, 
        new_row
    ])
 
    return result_2
```

```python
def draw_piechart(input_data,group,region):
    
    labels=input_data[group]
    sizes=input_data['sum']
    fig1, ax1 = plt.subplots()
    cmap=plt.get_cmap('PRGn_r')
    colors = cmap(
    np.linspace(
        0., 1., len(
            input_data[group]
        )))
    textprops = {"fontsize":12}
    ax1.pie(
        sizes, 
        labels=labels, 
        autopct='%1.1f%%',
        colors=colors,
        startangle=90,
        textprops = textprops
    )
    centre_circle = plt.Circle(
        (0,0),
        0.70,
        fc='white'
    )
    fig = plt.gcf()
    fig.gca().add_artist(centre_circle)
    ax1.axis('equal') 
    plt.title(
        'Доля продаж в {0} для {1}'.format(region,group), 
        fontsize = 15)
    plt.show()
```

### Влияние рейтинга ESRB на продажи по регионам

#### Подготовка функций

Для анализа влияния рейтинга на объем продаж подготовим фунцкии `group_rating` и `draw_barplot`.

```python
def group_rating(region):
    result = pd.DataFrame({
        'sum': data_actual.groupby([
            'rating'
        ])[region].agg('sum')
        }).reset_index()
    
    result=result.sort_values(
        by='sum',
        ascending = False
    ).reset_index()
    
    del(result['index'])
    
    return result
```

```python
def draw_barplot(input_data,region,region_name):
    
    fig = plt.figure(figsize=[12,3])
    ax = fig.add_axes([0,0,1,1])
    cmap = plt.get_cmap("PRGn")
    colors = cmap(
        np.linspace(
            0., 
            1.1,
            len(
                input_data['rating']
            )))
    rating = input_data['rating']
    values = input_data['sum']
    ax.bar(
        rating, 
        values,
        width = .3,
        color = colors
    )
    plt.title(
        'Диаграмма распределения объемов продаж в {0} по рейтингу ESRB'.format(region_name), 
        fontsize = 15
    )

    plt.xlabel('Рейтинг ESRB')
    plt.ylabel('Объем продаж, млн копий')
    plt.show()
```

## Проверка гипотез
### Сравнение оценки пользователей платформ Xbox One и PC

Сформулируем гипотезы для проверки.

<div style="border-radius: 15px;border:solid navy 1.5px; padding: 20px; box-shadow: 2px 2px 2px"> 

_**H₀:**_ Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

_**H₁:**_ Средние пользовательские рейтинги платформ Xbox One и PC различаются.   
    
    
</div>

При формулировании нулевой гипотезы исходим из того, что между событиями (в данном случае, средними оценками пользователей) нет взаимосвязи, и возможные совпадения являются случайными.

Выбор теста `st.ttest_ind` обусловлен тем, что выборки не зависимые, то есть в них нет переменных, которые бы измерялись дважды для одних и тех же единиц. 

Установим критическое значение `alpha` для p-value на уровне 5%. Если p-value окажется больше критического уровня, то нулевая гипотеза будет отвергнута, и альтернативная гипотеза будет принята верной.

Для того, чтобы указать значение `equal_var`, ответим на вопрос, является ли распределение нормальным. Для этого проведем тест Шапиро-Уилка.

```python
stat, p = shapiro(
    data_actual[
        data_actual['platform'] == 'XOne'
    ]['user_score'])
print(
    'Statistics=%.3f, p=%.3f' % (stat, p)
)

alpha = ALPHA
if p > alpha:
    print(
        'Нулевую гипотезу о нормальном распределении данных отвергнуть не удалось'
    )
else:
    print(
        'Нулевую гипотезу о нормальном распределении данных удалось отвергнуть, распределение не является нормальным'
    )
```

```python
stat, p = shapiro(
    data_actual[
        data_actual['platform'] == 'PC'
    ]['user_score'])
print(
    'Statistics=%.3f, p=%.3f' % (stat, p)
)

alpha = ALPHA
if p > alpha:
    print(
        'Нулевую гипотезу о нормальном распределении данных отвергнуть не удалось'
    )
else:
    print(
        'Нулевую гипотезу о нормальном распределении данных удалось отвергнуть, распределение не является нормальным'
    )
```

Для обеих выборок нулевая гипотеза о нормальном распределении была отвергнута, поэтому при проверке гипотезы укажем `equal_var = False`.

```python
alpha = ALPHA 

results = st.ttest_ind(
    data_actual[
        data_actual['platform'] == 'XOne'
    ]['user_score'], 
    data_actual[
        data_actual['platform'] == 'PC'
    ]['user_score'],
    equal_var = False)

if results.pvalue < alpha:
    print(
        "Отвергаем нулевую гипотезу"
    )
else:
    print(
        "Не получилось отвергнуть нулевую гипотезу"
    ) 
```

Значение p-value не превышает критический уровень 5%, следовательно отвергнуть нулевую гипотезу не удалось. Следовательно, у нас нет оснований утверждать, что средние пользовательские рейтинги платформ Xbox One и PC значимо различаются.
