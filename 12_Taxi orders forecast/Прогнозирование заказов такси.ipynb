Проанализируем сезонность данных за период **июль и август 2018**. Чтобы исключить возможные ошибочные выводы, включим в анализ два месяца.

```python
def draw_seasonal(data):
    decomposed = seasonal_decompose(data)

    plt.figure(figsize=(12, 12))
    plt.subplot(311)

    decomposed.trend.plot(ax=plt.gca())
    plt.title('Тренд')
    plt.subplot(312)
    decomposed.seasonal.plot(ax=plt.gca())
    plt.title('Сезонность')
    plt.subplot(313)
    decomposed.resid.plot(ax=plt.gca())
    plt.title('Шумы')
    plt.tight_layout()
```

#### Проверка стационарности ряда

Проверим, является ли временной ряд стационарным с помощью теста Дики-Фуллера.

```python
p_value = adfuller(
    data['num_orders'],
    regression='ctt'
)[1]

print('P =', p_value)

if p_value < 0.05:
    print(
        'Нулевая гипотеза отвергается. Ряд стационарен по критерию Дики-Фуллера'
    )
else:
    print(
        'Нулевая гипотеза не отвергается. Ряд нестационарен по критерию Дики-Фуллера'
    )
```

Подтверждается гипотеза о **стационарности временного ряда**, что свидетельсвует о том, что среднее, дисперсия и ковариация между периодами одинакового расстояния постоянны.

### Создание дополнительных признаков

Для построения модели дополним датасет дополнительными признаками. В данном случае добавим **день недели, час** и булев признак **выходной/будни** для каждой строки, а также **разности временного ряда** и **скользящее среднее**. 

```python
def make_features(data, max_lag, rolling_mean_size):

    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    data[
        'is_weekend'
    ] = data.index.dayofweek.isin([5,6])*1
    
    for lag in range(1, max_lag + 1):
        data[
            'lag_{}'.format(lag)
        ] = data['num_orders'].shift(lag)
    
    data['rolling_mean'] = data[
        'num_orders'
    ].shift().rolling(
        rolling_mean_size
    ).mean()
```

```python
make_features(data, 24, 12)
```

Удалим строки с отсутствующими значениями в столбцах для разности временного ряда.

```python
data = data.dropna()

# Проверка 
data.head(3)
```

После выбора модели проверим ее адекватность с помощью Dummy модели и рассчитает целевую метрику RMSE на тестовой выборке.

### Случайный лес

Найдем оптимальные параметры модели с помощью `GridSearchCV`.

```python
parametrs = { 
    'max_depth': range (8,12,1),
    'min_samples_leaf': range (1,3,1),
    'min_samples_split': range (1,4,1),
    'n_estimators': range (80,120,10)
}
```

```python
%%time

rf_reg = RandomForestRegressor()

rf_search = GridSearchCV(
    estimator=rf_reg, 
    cv=TSCV, 
    param_grid=parametrs, 
    scoring='neg_mean_squared_error', 
    n_jobs=-1, 
    verbose=1
)
rf_search.fit(
    features_train,
    target_train
)
```

```python
rf_search.best_params_
```

```python
print(
    'RMSE на обучающей выборке: ', 
    (abs(rf_search.best_score_)**0.5).round(2)
)
```

```python
reg_lr = LinearRegression()
rmse_lr = (
    abs(
        cross_val_score(
            reg_lr, 
            features_train, 
            target_train, 
            scoring="neg_mean_squared_error", 
            cv=TSCV
        ))**0.5).mean()

print(
    'RMSE на обучающей выборке:', 
    rmse_lr.round(2)
)
```

### Гребневая регрессия

Найдем оптимальные параметры модели с помощью `GridSearchCV`.

```python
parametrs = {
    'alpha':[
        0.001,0.01,0.1,1
    ],
    'solver':[
        'auto', 
        'svd', 
        'cholesky', 
        'lsqr',
        'sparse_cg'
    ],
    'max_iter': range(1,10,1)
}
```

```python
%%time

r_reg = Ridge()
tscv = TimeSeriesSplit(
    n_splits=10
)
r_search = GridSearchCV(
    estimator=r_reg, 
    cv=TSCV, 
    param_grid=parametrs, 
    scoring='neg_mean_squared_error', 
    n_jobs=-1, 
    verbose=1
)
r_search.fit(
    features_train, 
    target_train
)
```

Лучшее значение целевой метрики **RMSE** при использовании валидационной выборки достигнуто с использованием модели **Линейная регрессия**.
