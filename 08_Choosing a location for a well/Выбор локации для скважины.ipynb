## <font color = 'peru'>Описание проекта</font>

    
Добывающей компании **«ГлавРосГосНефть»** нужно решить, в каком регионе из исследуемых бурить новую скважину.
    
**Шаги для выбора локации** обычно такие:
 * В избранном регионе собирают характеристики для скважин: качество нефти и объём её запасов;
 * Строят модель для предсказания объёма запасов в новых скважинах;
 * Выбирают скважины с самыми высокими оценками значений;
 * Определяют регион с максимальной суммарной прибылью отобранных скважин.
    
Исследование проводится с использованием данных, содержащих информацию о пробах нефти в трёх регионах. Характеристики для каждой скважины в регионе уже известны. Необходимо построить модель для определения региона, где добыча принесёт наибольшую прибыль, и провести анализ возможной прибыли и рисков с использованием техники Bootstrap.
    
Данные геологоразведки трёх регионов находятся в файлах:
 * `/datasets/geo_data_0.csv`
 * `/datasets/geo_data_1.csv`
 * `/datasets/geo_data_2.csv`


### <font color = 'peru'>Условия задачи:</font>
 * Для обучения модели подходит только линейная регрессия (остальные — недостаточно предсказуемые).
 * При разведке региона исследуют 500 точек, из которых с помощью машинного обучения выбирают 200 лучших для разработки.
 * Бюджет на разработку скважин в регионе — 10 млрд рублей.
 * При нынешних ценах один баррель сырья приносит 450 рублей дохода. Доход с каждой единицы продукта составляет 450 тыс. рублей,  поскольку объём указан в тысячах баррелей.
 * После оценки рисков нужно оставить лишь те регионы, в которых вероятность убытков меньше 2.5%. Среди них выбирают регион с наибольшей средней прибылью.
 * Данные синтетические: детали контрактов и характеристики месторождений не разглашаются.
    
<h3 style="color:peru; margin-bottom:20px">Цель исследования </h3> 

**На основании анализа данных и с использованием техники Bootstrap, определить регион, бурение новой скважины в котором даст наибольшую прибыль с вероятностью риска убытков менее 2.5%.**

### Анализ распределения данных

Проанализируем показатели описательной статистики и картину распределения значений в датасетах с помощью диаграмм размаха и гистограмм. Для этого создадим функцию `describe_boxplot_hist`.

```python
def describe_boxplot_hist(df, input_column):

# Показатели описательной статистики
    print(
        df[input_column].describe()
    )

# Диаграмма размаха 
    flierprops = dict(
        marker='o', 
        markerfacecolor='None', 
        markersize=3,  
        markeredgecolor='black'
    )
    sns.set(
        rc={
            'figure.figsize':(16,1.5)
        })
    ax = sns.boxplot(
        x=df[input_column],
        showmeans=True,
        color ="sandybrown",
        width = .4,
        meanprops={
            "marker":"o",
            "markerfacecolor":"white", 
            "markeredgecolor":"black",
            "markersize":"8"},
        flierprops = flierprops
    )
    plt.title(
        'Диаграмма размаха для распределения значений {0}'.format(
            input_column.upper()), 
        fontsize=15
    )
    
# Гистограмма    
    fig= plt.subplots(
        figsize=(16,3)
    )
    sns.distplot(
        df[input_column], 
        color = 'peru'
    ).set_title(
        'Гистограмма размаха для распределения значений {0}'.format(
            input_column.upper()), 
        fontsize=15
    )
```

## <font color = 'peru'> Подготовка к расчёту прибыли </font>

Сохраним все ключевые значения для последующих расчётов в отдельных переменных.

```python
# Общий бюджет на разработку
BUDGET = 1e+10

# Стоимость единицы продукта (тыс. баррелей)
PRODUCT_UNIT_PRICE = 4.5e+5

# Планируемое количество скважин
POINTS = 200

# Количество исследуемых скважин
POINTS_STUDIED = 500

# Приемлемое значение вероятности убытков
ACCEPT_RISK = 0.025

# Количество экспериментов
EXP = 1000
```

## <font color = 'peru'> Расчёт прибыли и рисков </font>

Используя бутстрап, рассчитаем среднюю прибыль, доверительный интервал и риск убытков для каждого региона. Проведем 1000 экспериментов, в каждый из которых включим 500 скважин. 

```python
predictions = [
    predict_valid_0, 
    predict_valid_1, 
    predict_valid_2
]
targets_valid = [
    target_valid_0, 
    target_valid_1, 
    target_valid_2
]
```

```python
def profit_calc(
    target, 
    probabilities, 
    count
):
    
    probs_sorted = probabilities.sort_values(
        ascending=False
    )
    selected = target[probs_sorted.index][:count]
    return selected.sum() * PRODUCT_UNIT_PRICE - BUDGET

state = np.random.RandomState(42)

for i in range(len(targets_valid)):
    values = []
    counter = 0
    for _ in range(EXP):
        target_subsample = targets_valid[i].sample(
            n = POINTS_STUDIED, 
            replace=True, 
            random_state=state
        )
        probs_subsample = predictions[i][
            target_subsample.index
        ] 
        profit = profit_calc(
            target_subsample, 
            probs_subsample, 
            POINTS
        )
        values.append(profit)
        if profit < 0:
            counter += 1
            
    name = 'Регион ' + str(i)
    values = pd.Series(values)
    lower = values.quantile(ACCEPT_RISK)
    upper = values.quantile(1 - ACCEPT_RISK)
    mean =  values[
        (lower < values) & (values < upper)
    ].mean()

    print(name)
    print("Средняя прибыль: {:.2f}".format(mean))
    print (
        '95% доверительный интервал: {:.2f} - {:.2f}'.format(
            lower, 
            upper
        ))
    print('Риск убытков: {:.2%}'.format(counter / EXP))
    print('Количество убыточных экспериментов: ', counter)
    print('')
```

* По результатам экспериментов с помощью бутстрап приемлемый уровень риска убытков (менее 2.5%) получен только в **регионе 1**. 
* Полученная в ходе экспериментов **средняя прибыль** в регионе 1 составила **514 млрд рублей**. Количество убыточных экспериментов - 16.
