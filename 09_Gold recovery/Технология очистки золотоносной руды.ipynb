## Описание технологического процесса


Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

**Схема технологического процесса очистки:**

![gold_recovery](img/gold_recovery.jpg)


**1. Флотация**

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

**2. Очистка**


Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

##### Технологический процесс
 * **Rougher feed** — исходное сырье
 * **Rougher additions** (или **reagent additions**) — флотационные реагенты: Xanthate, Sulphate, Depressant
     * **Xanthate** — ксантогенат (промотер, или активатор флотации);
     * **Sulphate** — сульфат (на данном производстве сульфид натрия);
     * **Depressant** — депрессант (силикат натрия).
 * **Rougher process** (англ. «грубый процесс») — флотация
 * **Rougher tails** — отвальные хвосты
 * **Float banks** — флотационная установка
 * **Cleaner process** — очистка
 * **Rougher Au** — черновой концентрат золота
 * **Final Au** — финальный концентрат золота
 
 
##### Параметры этапов
 * **air amount** — объём воздуха
 * **fluid levels** — уровень жидкости
 * **feed size** — размер гранул сырья
 * **feed rate** — скорость подачи
 
 
#### Наименование признаков
Наименование признаков имеет формат:

`[этап].[тип_параметра].[название_параметра]`

_**Пример**: rougher.input.feed_ag_

_**Возможные значения для блока [этап]:**_
 * **rougher** — флотация
 * **primary_cleaner** — первичная очистка
 * **secondary_cleaner** — вторичная очистка
 * **final** — финальные характеристики
 
 
_**Возможные значения для блока [тип_параметра]:**_
 * **input** — параметры сырья
 * **output** — параметры продукта
 * **state** — параметры, характеризующие текущее состояние этапа
 * **calculation** — расчётные характеристики
 
 
Данные индексируются датой и временем получения информации (признак `date`). 

Соседние по времени параметры часто похожи.

Некоторые параметры недоступны, потому что замеряются и/или рассчитываются значительно позже. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков.

### Проверка расчета эффективности обогащения

Эффективность обогащения рассчитывается по формуле:

![formula](img/Recovery.jpg)

где:
* **C** — доля золота в концентрате после флотации/очистки;
* **F** — доля золота в сырье/концентрате до флотации/очистки;
* **T** — доля золота в отвальных хвостах после флотации/очистки.

Проанализируем значения эффективности обогащения в датасете, построив гистограмму и выведя показатели описательной статистики.

```python
def describe_boxplot_hist(df, input_column):

# Показатели описательной статистики
    print(df[input_column].describe())

# Диаграмма размаха 
    flierprops = dict(
        marker='o', 
        markerfacecolor='None', 
        markersize=3,  
        markeredgecolor='black'
    )
    sns.set(rc={'figure.figsize':(16,1.5)})
    ax = sns.boxplot(
        x=df[input_column],
        showmeans=True,
        color ="sandybrown",
        width = .4,
        meanprops={
            "marker":"o",
            "markerfacecolor":"white", 
            "markeredgecolor":"black",
            "markersize":"8"
        },
        flierprops = flierprops
    )
    plt.title(
        'Диаграмма размаха для распределения значений {0}'.format(
            input_column.upper()
        ), 
        fontsize=15
    )
    
# Гистограмма    
    fig= plt.subplots(figsize=(16,3))
    sns.distplot(
        df[input_column], 
        color = 'peru'
    ).set_title(
        'Гистограмма размаха для распределения значений {0}'.format(
            input_column.upper()
        ), 
        fontsize=15
    )
```

```python
describe_boxplot_hist(
    data_train, 
    'rougher.output.recovery'
)
```

Подготовим формулу для расчета эффективности обогащения из значений **C**, **T** и **F**.

```python
def recovery_calculation(df, C, T, F):
    C = df[C]
    T = df[T]
    F = df[F]
    recovery = (C * (F - T) * 100) / (F * (C - T))
    return recovery
```

Применим функцию `recovery_calculation()`.

```python
calculated_recovery = recovery_calculation(
    data_train,
    'rougher.output.concentrate_au', 
    'rougher.output.tail_au', 
    'rougher.input.feed_au'
)
```

Проверим общую картину распределения и значения описательной статистики для рассчитанной эффективности обогащения.

```python
# Показатели описательной статистики
print(calculated_recovery.describe())

# Диаграмма размаха 
flierprops = dict(
    marker='o', 
    markerfacecolor='None', 
    markersize=3,  
    markeredgecolor='black'
)
sns.set(rc={'figure.figsize':(16,1.5)})
ax = sns.boxplot(
    x=calculated_recovery,
    showmeans=True,
    color ="sandybrown",
    width = .4,
    meanprops={
        "marker":"o",
        "markerfacecolor":"white", 
        "markeredgecolor":"black",
        "markersize":"8"
    },
    flierprops = flierprops)
plt.title(
    'Диаграмма размаха для распределения значений Calculated recovery', 
    fontsize=15
)
    
# Гистограмма    
fig= plt.subplots(figsize=(16,3))
sns.distplot(
    calculated_recovery, 
    color = 'peru'
).set_title(
    'Гистограмма размаха для распределения значений Calculated recovery', 
    fontsize=15
);
```

Диаграмма размаха и гистограмма для исходных и расчетных значений очень похожи, параметры описательной статистики расходятся очень незначительно. Проверим корректность рассчетов, выведя значение среднего абсолютного отклонения (**MAE**)

#### Удаление выбросов

На графике изменения концентрации металлов по ходу процесса можно заметить пики около отметки **0**, которые являются выбросами. 

Удалим выбросы с помощью функции `delete_outliers()`, которая удалит данные со значениями **z-score** больше **5**.

```python
def delete_outliers(df):
    before = df.shape[0]
    df = df[(
        np.abs(stats.zscore(df)) < 5
    ).all(axis=1)]
    
    df = df.reset_index()
    after = df.shape[0]
    display(
        'Удалено {:.2%} строк'.format((
            before-after
        )/before
        ))
    return df
```

### Подготовка функции для расчета sMAPE

Для решения задачи будем использовать метрику качества — **sMAPE** (англ. _Symmetric Mean Absolute Percentage Error_, «симметричное среднее абсолютное процентное отклонение»).

Метрика sMAPE рассчитывается по следующей формуле:

![sMAPE](img/sMAPE.png)

Итоговая метрика sMAPE рассчитывается так:

![total_sMAPE](img/total_sMAPE.png)

Создадим функцию для расчета симметричного среднего абсолютного процентного отклонения (**_sMAPE_**) `smape_calculation` и добавим эту метрику в расчет оценки кросс-валидации.

```python
def smape_calculation(
    rougher_recovery_real, 
    rougher_recovery_predicted, 
    final_recovery_real, 
    final_recovery_predicted
): 
    rougher_smape = 1 / rougher_recovery_real.shape[0] * (((
        abs(
            rougher_recovery_predicted - rougher_recovery_real
        )) / ((
        abs(
            rougher_recovery_real
        ) + abs(
            rougher_recovery_predicted
        )) / 2)).sum()) * 100
    
    final_smape = 1 / final_recovery_real.shape[0] * (((
        abs(
            final_recovery_predicted - final_recovery_real
        )) / ((
        abs(
            final_recovery_real
        ) + abs(
            final_recovery_predicted
        )) / 2
    )).sum()) * 100
    
    total_smape = (
        rougher_smape * 0.25
    ) + (
        final_smape * 0.75
    )
    scorer = make_scorer(
        total_smape, 
        greater_is_better=False
    )
    
    return total_smape
```
